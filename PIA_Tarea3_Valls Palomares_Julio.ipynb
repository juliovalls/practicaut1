```python
# Función para calcular el precio final con IVA y descuento
def calcular_precio_final(precio_net, iva_porcentaje, descuento_porcentaje):
    precio_con_iva = precio_net * (1 + iva_porcentaje)
    precio_con_descuento = precio_con_iva * (1 - descuento_porcentaje)
    return precio_con_descuento

# Función para agregar un artículo al diccionario
def agregar_articulo(tienda, nombre, precio_net):
    if nombre in tienda:
        print("El artículo ya existe en la tienda.")
    else:
        tienda[nombre] = {"precio_net": precio_net, "cantidad": 0}
        print(f"Artículo '{nombre}' agregado a la tienda con precio neto de {precio_net:.2f}.")

# Función para consultar todos los artículos en el diccionario
def consultar_articulos(tienda, iva_porcentaje, descuento_porcentaje):
    if not tienda:
        print("El diccionario de la tienda está vacío.")
    else:
        print("Información de los artículos:")
        for nombre, info in tienda.items():
            precio_net = info["precio_net"]
            cantidad = info["cantidad"]
            precio_iva = precio_net * (1 + iva_porcentaje)
            precio_final = calcular_precio_final(precio_net, iva_porcentaje, descuento_porcentaje)
            print(f"{nombre} - Precio neto: {precio_net:.2f} - Precio con IVA: {precio_iva:.2f} - Precio con descuento: {precio_final:.2f} - Cantidad: {cantidad}")

# Función para consultar el PVP de un artículo en el diccionario
def consultar_pvp(tienda, nombre, iva_porcentaje, descuento_porcentaje):
    if nombre in tienda:
        info = tienda[nombre]
        precio_net = info["precio_net"]
        precio_iva = precio_net * (1 + iva_porcentaje)
        precio_final = calcular_precio_final(precio_net, iva_porcentaje, descuento_porcentaje)
        print(f"El PVP del artículo '{nombre}' es: {precio_final:.2f} (Precio neto: {precio_net:.2f}, Precio con IVA: {precio_iva:.2f})")
    else:
        print(f"El artículo '{nombre}' no está en la tienda.")

# Función para modificar el precio de un artículo en el diccionario
def modificar_precio(tienda, nombre, nuevo_precio_net):
    if nombre in tienda:
        tienda[nombre]["precio_net"] = nuevo_precio_net
        print(f"El precio del artículo '{nombre}' ha sido modificado a {nuevo_precio_net:.2f}.")
    else:
        print(f"El artículo '{nombre}' no está en la tienda.")

# Función para borrar un artículo del diccionario
def borrar_articulo(tienda, nombre):
    if nombre in tienda:
        del tienda[nombre]
        print(f"El artículo '{nombre}' ha sido eliminado de la tienda.")
    else:
        print(f"El artículo '{nombre}' no está en la tienda.")

# Función para realizar acciones del menú
def menu():
    tienda = {
        "Camiseta": {"precio_net": 20, "cantidad": 100},
        "Zapatos deportivos": {"precio_net": 60, "cantidad": 50},
        "Gafas de sol": {"precio_net": 25, "cantidad": 30},
        "Teléfono inteligente": {"precio_net": 400, "cantidad": 10}
    }

    iva_porcentaje = 0.21
    descuento_porcentaje = 0.10

    while True:
        print("\nMenú:")
        print("1. Mostrar lista de artículos")
        print("2. Consultar PVP de un artículo")
        print("3. Añadir nuevo artículo")
        print("4. Modificar precio de un artículo")
        print("5. Borrar un artículo")
        print("6. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == '1':
            consultar_articulos(tienda, iva_porcentaje, descuento_porcentaje)
        elif opcion == '2':
            nombre = input("Ingrese el nombre del artículo para consultar el PVP: ")
            consultar_pvp(tienda, nombre, iva_porcentaje, descuento_porcentaje)
        elif opcion == '3':
            nombre = input("Nombre del nuevo artículo: ")
            while True:
                try:
                    precio_net = float(input("Precio neto del nuevo artículo: "))
                    break
                except ValueError:
                    print("Por favor, ingrese un número válido para el precio neto.")
            agregar_articulo(tienda, nombre, precio_net)
            print("Artículo agregado correctamente.")
        elif opcion == '4':
            nombre = input("Ingrese el nombre del artículo para modificar el precio: ")
            while True:
                try:
                    nuevo_precio_net = float(input("Nuevo precio neto: "))
                    break
                except ValueError:
                    print("Por favor, ingrese un número válido para el nuevo precio neto.")
            modificar_precio(tienda, nombre, nuevo_precio_net)
        elif opcion == '5':
            nombre = input("Ingrese el nombre del artículo para borrar: ")
            borrar_articulo(tienda, nombre)
        elif opcion == '6':
            break
        else:
            print("Opción no válida. Inténtelo de nuevo.")

# Ejecutar el menú
menu()
```
